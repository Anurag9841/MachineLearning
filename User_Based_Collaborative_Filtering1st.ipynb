```python
def RMSE(y_true,y_predicted):
  return np.sqrt(mean_squared_error(y_true,y_predicted))
```

```python
def recommend_song(user_id,k=10,thres_hold=0.2):
  # Similarity_score
  similar_user=similarity_df[similarity_df[user_id]>thres_hold][user_id].sort_values(ascending=False)[:k]
  avg_user_rating = rating_matrix.T.mean()[user_id]
  # Song_listend By user
  seen_df = rating_matrix[rating_matrix.index == user_id].dropna(axis=1, how='all')
  similar_user_songs = rating_matrix[rating_matrix.index.isin(similar_user.index)].dropna(axis=1, how='all')
  #Drop Song_Listened_By_user
  similar_user_songs.drop(seen_df.columns, axis=1, inplace=True, errors="ignore")
  item_score = {}
  for item in similar_user_songs.columns:
      songs_rating = similar_user_songs[item]
      total_score = 0
      count=0
      for v in similar_user.index:
          if not pds.isna(songs_rating[v]):
              score = songs_rating[v] * similar_user[v]
              total_score += score
              count +=1
      item_score[item] = total_score / count
  item_score=pds.DataFrame(item_score.items(),columns=["Song","song_score"])
  ranked_item_score=item_score.sort_values(by="song_score",ascending=False)
  return ranked_item_score.head(10)
```
